```python
#unsupervised learning


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans


#load dataset
data = pd.read_csv('students_dropout_academic_success.csv')
data
```

### DATA PREPROCESSING AND EXPLORATION

```python
data.isnull().sum()
data.duplicated().sum()
```

```python
data.columns = ['Marital_status','Application_Mode','Application_Order','Course','Attendance','Qualification',
                'Qualification_Grade','Nationality','Qualification_M','Qualification_F','Occupation_M','Occupation_F',
                'Admission_Grade','Displaced','Education_special_needs','Debtor','Fees_Paid','Gender','Scholarship','Age','International',
                'Semester1_Credited','Semester1_Enrolled','Semester1_Evaluations','Semester1_Approved',
                'Semester1_Grade','Semester1_Without_evaluations','Semester2_Credited','Semester2_Enrolled',
                'Semester2_Evaluaions','Semester2_Approved','Semester2_Grade','Semeter2_without_evaluations',
                'Unemployment_rate','Inflation_rate','GDP','target']
                
                
data.target = (data.target == 'Dropout').astype(int)
data.head()
```

```python
#Removing Outliers Using Interquartile Range of Age Feature
ages = data['Age']
q1 = ages.quantile(0.25)
q3 = ages.quantile(0.75)
iqr = q3 = q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
data_cleaned = data[(ages >= lower_bound) & (ages <= 55)]#Recting upper bound to avooid loss of mean

#Removing Outliers From Semester1_Enrolled Feature
data_cleaned = data_cleaned[data_cleaned.Semester1_Enrolled < 20]
data_cleaned


```

```python
#pairwise plot of Selected Features
#data_without_enrolled = data[data.target != 2]
sns.set(rc= {'axes.facecolor': '#FFF9ED','figure.facecolor': '#FFF9ED'})
#PALLET = ['#']
to_plot = ['Age', 'Admission_Grade', 'Semester1_Enrolled', 'Semester1_Approved', 'Semester1_Grade', 'target']
print('Relative plot of selected Features')
sns.pairplot(data_cleaned[to_plot], hue= 'target', palette= ['#682F2F', '#F3AB60'])
plt.show()
```

```python
#Hatmap to show correlation of Features
from matplotlib.colors import ListedColormap

cmap = ListedColormap(['#682F2F','#9E726F','#D6B2B1','#B9C0C9','#9F8A78'])
corrmat = round(data_cleaned.corr(), 2)
plt.figure(figsize= (20, 20))
sns.heatmap(corrmat, annot= True, cmap= cmap, center= 0)
```

```python
# Scaling Dataset by Standardization

target = data_cleaned.target
del data_cleaned['target'] # Removing target feature

scaler = StandardScaler()
scaler.fit(data_cleaned)
scaled_data_cleaned = pd.DataFrame (scaler.transform(data_cleaned), columns = data_cleaned.columns)
scaled_data_cleaned
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components= 3)
pca.fit(scaled_data_cleaned)
pca_data_cleaned = pd.DataFrame(pca.transform(scaled_data_cleaned),
                    columns= ['col1', 'col2', 'col3'])
pca_data_cleaned.describe().T
```

```python
#pca_data_cleaned.describe().T
```

```python
x = pca_data_cleaned['col1']
y = pca_data_cleaned['col2']
z = pca_data_cleaned['col3']
fig = plt.figure(figsize= (10,8))
ax = fig.add_subplot(111, projection= '3d')
ax.scatter(x,y,z, c= 'maroon', marker= 'o')
ax.set_title('3D Projection of Reduced Dimensions')
plt.show()
```

```python
#using the kelbow visualizer to know the number of clusters the dataset will form

from yellowbrick.cluster import KElbowVisualizer

print ('Elbow method to determine number of clusters')
elbow_m = KElbowVisualizer(KMeans(), k=10)
elbow_m.fit(pca_data_cleaned)
elbow_m.show()
```

```python
# Clustering

kmeans = KMeans(n_clusters= 4)
predictions = kmeans.fit_predict(pca_data_cleaned)
labels = kmeans.labels_
centers = kmeans.cluster_centers_

pca_data_cleaned['Clusters'] = predictions
data_cleaned['Clusters'] = predictions

# to visualize the clusters

fig = plt.figure(figsize= (10,8))
ax = plt.subplot(111, projection= '3d', label= 'bla')
ax.scatter (x, y, z, s= 40, c= pca_data_cleaned['Clusters'], marker= 'o', cmap = cmap)
ax.set_title('The Plot of the Clusters')
plt.show()
```

```python
# to create a table that shows how the target and clusters correlate

pd.crosstab(data_cleaned['Clusters'], target, normalize = 'index')
```

```python
# to predict the dropout using their grade and number of courses passed

data_cleaned.groupby('Clusters')[[
    'Semester1_Approved',
    'Semester1_Grade',
    'Semester2_Approved',
    'Semester2_Grade'
]].mean()
```

```python
# using other columns to predict the dropout;
# to show the relationshio between clusters

data_cleaned.groupby('Clusters')[[
    'Semester1_Approved',
    'Semester1_Enrolled',
    'Semester2_Approved',
    'Semester2_Enrolled'
]].mean()
```
